```python
import pandas as pd
import numpy as np
```

```python
def subsample_data(data, samp_proportion):
  """
  Create a random subset of the data
  
  params:
    - data: Data to make a subset of
    - samp_proportion: Amount of data to keep

  returns:
    - subset: subset of the data
  """
  size = len(data)
  samp_proportion = 0.7
  indices = np.array(range(0, size))

  rand_sample = np.random.choice(indices, size=int(size*samp_proportion), replace=False)
  subset = data.loc[rand_sample, :]
  return subset
```

```python
# Create a subset of the data: This will be used with the TinyHGR data
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/HandGestures/data/LeapMotion/all_data.csv")
```

```python
data_subset = subsample_data(data, 0.7)
```
